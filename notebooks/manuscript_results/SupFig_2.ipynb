```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For all scout data
data_dir = '/Users/kathleenkanaley/Desktop/grapes_from_space/data/'
scout_20 = pd.read_csv(data_dir +'scout/scout_incidence/scout_inc_2020.csv')
scout_21 = pd.read_csv(data_dir +'scout/scout_incidence/scout_inc_2021.csv')
scout_22 = pd.read_csv(data_dir +'scout/scout_incidence/scout_inc_2022.csv')

dfs = [scout_20, scout_21, scout_22]
```

```python
scout_20_nosept = scout_20[scout_20['Date']!='2020-09-09']
scout_21_noaug3 = scout_21[scout_21['Date']!='2021-08-03']
```

```python
dfs = [scout_20_nosept, scout_21_noaug3, scout_22]
```

```python
# For each year, plot histograms of GDM incidence and severity distributions

for df in dfs:
    ax=sns.displot(df, x="DM_severity", col="Date", binwidth=5, height=5)
    plt.xlabel('GDM Severity (%)')
    #plt.xlabel('x-axis label')
    plt.show()
```

```python
# 2020
plt_2020 = sns.displot(
    data=scout_20_nosept, x="DM_severity", col="Date", binwidth=1, height=5, color='black')
plt_2020.set_axis_labels("GDM severity (%)", "Count")
```

```python
# 2020
plt_2020_inc = sns.displot(
    data=scout_20_nosept, x="DM_inc", col="Date", binwidth=1, height=5, color='black')
plt_2020_inc.set_axis_labels("GDM incidence (%)", "Count")
```

```python
# 2020 severity - aggregate for whole year
sns.set_theme(style="darkgrid")
plt_2020agg = sns.displot(
    data=scout_20_nosept, x="DM_severity", binwidth=2, height=5, color='black').set(title='GDM Severity 2020',
                                                                                   xlim=(-5,105))
plt_2020agg.set_axis_labels("GDM severity (%)", "Count")
```

```python
# 2020 incidence - aggregate for whole year

plt_2020incagg = sns.displot(
    data=scout_20_nosept, x="DM_inc", binwidth=2, height=5, color='black').set(title='GDM Incidence 2020')
plt_2020incagg.set_axis_labels("GDM incidence (%)", "Count")
```

```python
scout_22['GDM_sev']='low'
scout_22.loc[scout_22['DM_severity'] > 10, 'GDM_sev'] = 'high'
    
scout_22['GDM_inc']='low'
scout_22.loc[scout_22['DM_inc'] > 25, 'GDM_inc'] = 'high'
```

```python
# Final SkySat image date in 2022: 2022-07-20
# Final PScope image date in 2022: 2022-08-02

# Ratio low:high severity on 2022-07-20

jul20 = scout_22[scout_22['Date']=='2022-07-20']

sev_count_jul20 = jul20.GDM_sev.value_counts()
print('Low severity 2022-07-20:', sev_count_jul20[0])
print('High severity 2022-07-20:', sev_count_jul20[1])
print('Proportion:', round(sev_count_jul20[0] / sev_count_jul20[1], 2), ': 1')

# # Ratio low:high inc on 2022-07-20
inc_count_jul20 = jul20.GDM_inc.value_counts()
print('Low incidence 2022-07-20:', inc_count_jul20[0])
print('High incidence 2022-07-20:', inc_count_jul20[1])
print('Proportion:', round(inc_count_jul20[0] / inc_count_jul20[1], 2), ': 1')
```

```python
# Final PScope image date in 2022: 2022-08-02

# Ratio low:high severity on 2022-08-02

aug02 = scout_22[scout_22['Date']=='2022-08-02']

sev_count_aug02 = aug02.GDM_sev.value_counts()
print('Low severity 2022-08-02:', sev_count_aug02[0])
print('High severity 2022-08-02:', sev_count_aug02[1])
print('Proportion:', round(sev_count_aug02[0] / sev_count_aug02[1], 2), ': 1')

# # Ratio low:high inc on 2022-07-20
inc_count_aug02 = aug02.GDM_inc.value_counts()
print('Low severity 2022-08-02:', inc_count_aug02[0])
print('High severity 2022-08-02:', inc_count_aug02[1])
print('Proportion:', round(inc_count_aug02[0] / inc_count_aug02[1], 2), ': 1')
```
