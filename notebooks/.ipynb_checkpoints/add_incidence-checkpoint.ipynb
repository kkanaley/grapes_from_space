```python
# Function c/o: https://stackoverflow.com/questions/9507648/datetime-from-string-in-python-best-guessing-string-format

def get_date(s_date):
    date_patterns = ["%m/%d/%Y", "%m/%d/%y"]

    for pattern in date_patterns:
        try:
            return str('2022-')+ datetime.strptime(s_date, pattern).strftime('%m-%d')
        except:
            pass
```

```python
pm_inc_2022['Date']=pm_inc_2022['Date'].apply(get_date)
dm_inc_2022['Date']=dm_inc_2022['Date'].apply(get_date)
```
