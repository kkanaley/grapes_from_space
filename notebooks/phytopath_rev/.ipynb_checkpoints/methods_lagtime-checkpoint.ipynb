```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Distribution of lag time between satellite image acquistion and observation (scout)

```python
# open lag data

lag_data = pd.read_csv('/Users/kathleenkanaley/Desktop/satacq_lagtimes.csv')

plt.rcParams["font.family"] = "Arial"

ax = sns.histplot(data=lag_data, x='lag', discrete="True", shrink=.8)
ax.set(xlabel="Lag (days)",
       ylabel = "Count (images)",
       ylim = (0,10),
       title='Lag between satellite acquisition and scout observation')
plt.tight_layout()
plt.savefig('/Users/kathleenkanaley/Desktop/lag_all.jpg')
plt.show()
```

```python
ps_data = lag_data[lag_data['source']=='PS']

ax = sns.histplot(data=ps_data, x='lag', discrete="True", shrink=.8)
ax.set(xlabel="Lag (days)",
       ylabel = "Count (images)",
       ylim = (0,10),
       title='Lag time\nPlanetScope acquisitions - Scout observations')
plt.tight_layout()
plt.savefig('/Users/kathleenkanaley/Desktop/lag_ps.jpg')
plt.show()
```

```python
ss_data = lag_data[lag_data['source']=='SS']

ax = sns.histplot(data=ss_data, x='lag', discrete="True", shrink=.8)
ax.set(xlabel="Lag (days)",
       ylabel = "Count (images)",
       ylim = (0,10),
       xlim = (-2.5,2.5),
       title='Lag time\nSkySat acquisitions - Scout observations')
plt.tight_layout()
plt.savefig('/Users/kathleenkanaley/Desktop/lag_ss.jpg')
plt.show()
```
